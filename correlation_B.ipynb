```python
import yfinance as yf
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

data = yf.download("PHYS-U.TO SIVR PPLT PPLT BTC ABX.TO FNV.TO WPM.TO PAAS.TO GDX GDXJ NEM RIO SAND SBSW GOLD CCO.TO CNQ.TO SU.TO ENB.TO URNM GUNR EOG CCJ SHEL TTE SLYV ADBE MSFT AMZN NVDA AAPL GOOGL ILF INDA VGK VPL VWO L.TO MRU.TO RY.TO BMO.TO TD.TO DOL.TO CTC-A.TO O", start='2002-01-01') 
x = data['Close'].pct_change()
corr = x.corr()

#sm.graphics.plot_corr(corr, xnames=list(x.columns))

fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            cmap="viridis")

#plt.savefig('plot.png')
```
